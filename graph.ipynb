```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib as mpl
import seaborn as sns
import os
%matplotlib inline


```

```python
auto=pd.read_csv("titanic.csv")
auto.head()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(auto['age'].dropna(), bins=30, edgecolor='blue')
plt.title('Distribution of Ages')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='pclass', data=auto)
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(auto['age'], auto['fare'], alpha=0.5)
plt.title('Age vs. Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.show()
```

```python
class_vs_sex = auto.groupby(["class", "sex"]).size().unstack(fill_value=0)

plt.figure(figsize=(12, 8))
class_vs_sex.plot(kind='bar', stacked=True)
plt.title('Passenger Class vs Sex')
plt.xlabel('Class')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```
